### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
# 1. Asking temperature for the 5 days
temperatures = []
for day in range(1, 6):
    temp = float(input(f"Enter temperature for Day {day}: "))
    temperatures.append(temp)

# 2. Writing file in format "Day 1: 22°C"
with open("temps.txt", "w") as f:
    for day, temp in enumerate(temperatures, start=1):
        f.write(f"Day {day}: {temp}°C\n")

# 3. Read file and count average temperature
with open("temps.txt", "r") as f:
    lines = f.readlines()

total = 0
count = 0

for line in lines:
    # line, example: "Day 1: 22.5°C\n"
    parts = line.split(":")        # ["Day 1", " 22.5°C\n"]
    temp_part = parts[1].strip()   # "22.5°C"
    temp_str = temp_part.replace("°C", "")  # "22.5"
    temp_value = float(temp_str)
    total += temp_value
    count += 1

average = total / count

print(f"Average temperature over {count} days: {average:.2f}°C")
`````

### Activity 9: Finding "Dream" Sentences in Les_Brown.txt

From the `Les_Brown.txt` file, read the content and identify all sentences that contain the word "dream". Write these sentences to a new file called `dream_sentences.txt`.

**Steps:**
1. Open and read the content of `Les_Brown.txt`.
2. Split the text into sentences.
3. Filter out sentences that contain the word "dream" (case-insensitive).
4. Write these sentences into the new file `dream_sentences.txt`.
<hr>

`````python
# 1. Read file Les_Brown.txt
with open("Les_Brown.txt", "r") as f:
    text = f.read()

# 2. Conent the text
# Splitting text.
sentences = text.split(".")

# 3. Identify all sentences that contain the word  "dream" 
dream_sentences = []
for s in sentences:
    if "dream" in s.lower():      # search "dream" in lower case
        cleaned = s.strip()       # remove spaces
        if cleaned:               # ignoring emptyy strings
            dream_sentences.append(cleaned + ".")  # place '.' back

# 4. Adding all in to the new file dream_sentences.txt
with open("dream_sentences.txt", "w") as out:
    for sentence in dream_sentences:
        out.write(sentence + "\n")

print("Sentences with 'dream' were saved to dream_sentences.txt")
`````

### Activity 13: Identifying Duplicate Entries in Clothing Stock

Create a program that reads the `clothing_stock.txt` file and identifies if there are any duplicate entries (both item and size). Use a set to find and print these duplicates.

**Steps:**
1. Read the file and store each item in a set.
2. Keep track of duplicate entries.
3. Print any duplicates found.
<hr>

`````python
# Duplicates in clothing_stock.txt
seen = set()         # store here
duplicates = set()   # duplicates

with open("clothing_stock.txt", "r") as f:
    for line in f:
        item = line.strip()      # remove \n и paces
        if item in seen:
            duplicates.add(item) # if repeated — this duplicate
        else:
            seen.add(item)

# Print results
if duplicates:
    print("Duplicate entries found:")
    for d in duplicates:
        print(d)
else:
    print("No duplicates found.")
`````

### Activity 14: Counting Different Sizes per Clothing Item

Create a Python program that reads data from the `clothing_stock.txt` file and counts how many different sizes of each clothing item are available. Store this in a dictionary and print it.

**Steps:**
1. Read the file and extract both clothing items and sizes.
2. Count how many different sizes exist for each item.
3. Store the results in a dictionary and print it.
<hr>

`````python
# Counting different sizes per clothing item
sizes_per_item = {}   # list: { "T-shirt": {"S", "M", "L"}, ... }

with open("clothing_stock.txt", "r") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        # split str to "item" and "size"
        parts = line.split(",")
        item = parts[0].strip()
        size = parts[1].strip()

        # adding size to the items
        if item not in sizes_per_item:
            sizes_per_item[item] = set()
        sizes_per_item[item].add(size)

# Printing, how many different sizes have the item
for item, sizes in sizes_per_item.items():
    print(f"{item}: {len(sizes)} different size(s) -> {sizes}")
`````

### Activity 19 (Set): Splitting Clothing Stock by Size

Create a Python program that reads the `clothing_stock.txt` file and splits the clothing items into separate sets based on their size (e.g., all XS items in one set, all M items in another set). Print each set.

**Steps:**
1. Read the file and separate items by their size.
2. Store each size in a different set.
3. Print each set of items.
<hr>

`````python
# Elizabeth Ruda
# M01289571
# Splitting Clothing Stock by Size

# Import library pandas for working with DataFrame
import pandas as pd
import os

# Creating variables for the file with clothing
data_file = "clothing_stock.csv"     # name of the CSV file with clothing data
size_order = ["XS", "S", "M", "L", "XL", "XXL"]  # list of clothing sizes in order

# Create a sample CSV file if it doesn't exist
if not os.path.exists(data_file):
    sample = pd.DataFrame({
        "Item": ["T-shirt", "Jeans", "Skirt", "Jacket", "Coat", "Top"],
        "Size": ["M", "L", "S", "XL", "XXL", "XS"]
    })
    sample.to_csv(data_file, index=False)
    print("Sample CSV file created.")


# Main function
def split_clothing_by_size_pandas(filename: str) -> pd.DataFrame:
    try:
        # Read CSV
        df = pd.read_csv(filename)

        # Validate required columns
        required = {"Item", "Size"}
        if not required.issubset(df.columns):
            raise ValueError("Input file must contain 'Item' and 'Size' columns.")

        # Clean column values
        df["Size"] = df["Size"].astype(str).str.strip().str.upper()

        # Keep only known sizes
        mask = df["Size"].isin(size_order)
        if df[~mask].any().any():
            raise ValueError(f"Unknown size values found: {df[~mask]['Size'].tolist()}")

        # Convert Size to a categorical type (to keep XS..XXL in order)
        df["Size"] = pd.Categorical(df["Size"], categories=size_order, ordered=True)

        # Group by Item and join sizes
        grouped = (
            df.groupby("Item", as_index=False, observed=True)["Size"]
            .apply(lambda xs: ", ".join(xs) if isinstance(xs, list) else "")
        )

        # Styling for HTML output
        styler = grouped.style.set_properties(
            **{
                "text-align": "left",
                "padding": "6px 10px"
            }
        ).set_table_styles(
            [
                {"selector": "th", "props": [("text-align", "left"), ("padding", "6px 10px")]},
                {"selector": "caption", "props": [("caption-side", "top"), ("font-weight", "600")]}
            ]
        )

        # Save CSV
        grouped.to_csv("clothing_by_size.csv", index=False)

        # Save HTML
        with open("clothing_by_size.html", "w", encoding="utf-8") as f:
            f.write(styler.to_html())

        print("\n✔ Results saved as 'clothing_by_size.csv' and 'clothing_by_size.html'")

        return grouped

    except FileNotFoundError:
        print("❌ Error: The file was not found.")
    except ValueError as e:
        print(f"❌ Data Error: {e}")
    except Exception as e:
        print(f"❌ Unexpected error: {e}")


# Run the function
result = split_clothing_by_size_pandas(data_file)
`````
